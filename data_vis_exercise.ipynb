```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cocoa_df = pd.read_excel("/content/drive/MyDrive/Coding Dojo/Week 3/cocoa beans/Global import of cocoa beans.xlsx")
```

```python
year_lis = (2015, 2016, 2017, 2018, 2019)

asia = cocoa_df.loc[cocoa_df["Region"] == "Asia & Oceania", :]
europe = cocoa_df.loc[cocoa_df["Region"] == "Europe", :]
america = cocoa_df.loc[cocoa_df["Region"] == "Americas", :]

plt.plot(year_lis, asia["Import in tonnes"].tolist(), label= "Asia & Oceania");
plt.plot(year_lis, europe["Import in tonnes"].tolist(), label= "Europe");
plt.plot(year_lis, america["Import in tonnes"].tolist(), label= "Americas");
plt.legend();
plt.xlabel("Year", fontsize= 15);
plt.ylabel("Tonnes of Cocoa", fontsize= 15);
plt.title("Global Import of Cocoa Beans", fontsize= 16);
plt.xlim(left= 2015, right= 2019);
plt.grid()
```
